```python
def cluster_name(prev, this, lev):
    _prev_path = prev.split('_')[1]
    n = str(lev)+'_'
    if this >= 0:
        n += str(_prev_path) + '-' + str(this)
    else:
        n += str(_prev_path)
    return n

def add_level_names(b):
    b['level_0'] = b.apply(lambda x: '0_' + str(x['cluster_r0']), axis=1)
    for i in range(1,11):
        b['level_'+str(i)] = b.apply(lambda x: cluster_name(x['level_'+str(i-1)], x['cluster_r'+str(i)], i), axis=1)
```

```python
def collect_nodes(b, levels):
    _seen = set([])
    for lvl in levels:
        #print lvl
        for node_name, group in b.groupby(lvl):
            #print '...'+node_name
            if (group.shape[0] >= 1000) and (node_name not in _seen):
                _seen.add(node_name)
                _doc_pivot = {'size': group.shape[0],
                              'keywords': lookup_keywords(node_name),
                              'id': node_name}
                outloc = '../cluster_viz/tree_data/'+node_name+'.json'
                with open(outloc, 'w') as doc_data_out:
                    _tree_data = {'id': node_name,
                                  'items': lookup_docs(group['doc_id'])}
                    json.dump(_tree_data, doc_data_out)
                yield _doc_pivot
```

```python
with open('../persistence/kmeans_clustered_docs.json','w') as fout:
    all_nodes = [node for node in collect_nodes(bookie, levels)]
    json.dump(all_nodes, fout)
```

```python
def reduce_key(k):
    level, path = k.split('_')
    parent_level = str(int(level) - 1)
    parent_path = '-'.join(path.split('-')[:-1])
    reduced = '_'.join([parent_level, parent_path])
    return reduced


def add_level(lvl, parent_key):
    nodes = filter(lambda x: reduce_key(x['id']) == parent_key, all_nodes)
    next_lvl = lvl + 1
    for node in nodes:
        node['children'] = add_level(next_lvl, node['id'])
    return nodes

root_nodes = ["0_0", "0_1", "0_2", "0_3"]
    
tree = filter(lambda x: x['id'] in root_nodes, all_nodes)

depth = 10

for root_node in tree:
    root_node['children'] = add_level(1, root_node['id']) 
```
